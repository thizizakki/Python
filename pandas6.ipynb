```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "City":["Chicago","New York","Orlando"],
    "Temperature": [21,14,35]
})
df1
```

```python
df2 = pd.DataFrame({
    "City":["Chicago","New York","Orlando"],
    "Humidity": [65,68,75]
},)
df2
```

```python
df3 = pd.merge(df1,df2,on='City')
df3
```

```python
df1 = pd.DataFrame({
    "City":["Chicago","New York","Orlando","Baltimore"],
    "Temperature": [21,14,35,32]
})
df1
```

```python
df2 = pd.DataFrame({
    "City":["Chicago","New York","San Fransisco"],
    "Humidity": [65,68,71]
},)
df2
```

```python
df3 = pd.merge(df1,df2,on='City') # Only matching cities will be fetched
df3
```

```python
df3 = pd.merge(df1,df2,on='City',how='outer') # Kind of union
df3
```

```python
df3 = pd.merge(df1,df2,on='City',how='left') #Left join 
df3
```

```python
df3 = pd.merge(df1,df2,on='City',how='right') #Right join 
df3
```

```python
df3 = pd.merge(df1,df2,on='City',how='outer',indicator=True) # # to indicate the source of each row: 'left_only', 'right_only', or 'both'.
df3
```

```python
#Suffixes
df1 = pd.DataFrame({
    "City":["Chicago","New York","Orlando","Baltimore"],
    "Temperature": [21,14,35,32],
    "Humidity": [65,68,71,75]
})
df1
```

```python
df2 = pd.DataFrame({
    "City":["Chicago","New York","San Diego"],
    "Temperature": [21,14,35],
    "Humidity": [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1,df2,on='City',suffixes=('_left','_right'))
df3
```
