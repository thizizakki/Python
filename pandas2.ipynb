```python
#Creating a DataFrame from json
weather_data = {
    'day': ['1/1/2017','1/2/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017'],
    'event': ['Rain','Sunny','Snow','Snow','Rain','Sunny'],
    'temperature': [32,35,38,32,30,31],
    'windspeed': [6,7,2,7,4,2]
}

df = pd.DataFrame(weather_data)
df
```

```python
#creating a dataframe from the tuple
weather_data = {
    ('1/1/2017',32,6,'Rain'),
    ('1/2/2017',35,7,'Sunny'),
    ('1/3/2017',28,2,'Snow')
}

df = pd.DataFrame(weather_data, columns=['day','temperature','windspeed','event'])
df
```

```python
#creating dataframe from the dictionary
weather_data = [
    {'day':'1/1/2017','temperature':32,'windspeed':6,'event':'Rain'},
    {'day':'1/2/2017','temperature':35,'windspeed':7,'event':'Sunny'},
    {'day':'1/3/2017','temperature':28,'windspeed':2,'event':'Snow'},
]
df = pd.DataFrame(weather_data)
df
```

```python
def convert_people_cell(cell):
    if cell == "":
        return 'Sam Walton'
    return cell
df = pd.read_excel('new_without_header.xlsx',header=None,names=['tickers', 'eps', 'revenue', 'price', 'people'],converters={
    'people':convert_people_cell
})
df
```

```python
#Creates an excel file where the data starts with row 1 and column 2
df.to_excel("new_without_header1.xlsx",sheet_name="Stocks",startrow=1,startcol=2)
```

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL','WMT','MSFT'],
    'price': [845,65,64],
    'pe' : [30.37,14.26,30.97],
    'eps': [27.82,4.61,2.12]
})

df_weather = pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [25,32,8],
    'event': ['Rain','Sunny','Snow']
})
```
