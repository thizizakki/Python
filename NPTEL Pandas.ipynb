```python
#Grouped Means
#For each year in our data, what was the average life expectancy?
#To answer this question,
#we need to split our data into parts by year
#then we get the lifeExp column and calculate the mean

df.groupby('year')['lifeExp'].mean()
```

```python
#Multigroup
df.groupby(['year','continent'])[['lifeExp','gdpPercap']].mean()
```

```python
#To flatten the stacked table
df.groupby(['year','continent'])[['lifeExp','gdpPercap']].mean().reset_index()
```

```python
#Grouped frequency counts
#Use the nunique to get counts of unique values of a Pandas series

df.groupby('continent')['country'].nunique()
```

```python
#LifeExp overall 
df.groupby('year')['lifeExp'].mean()
```
